```python

from tabulate import tabulate
Newsdays = []
probability = []
cmltv_prob = []
temp = 0
n=int(input("Number of Newsday : "))

i=0
for i in range(n):
   newsday = input(f"Enter the newsday {i+ 1}: ")
   Newsdays.append(newsday)

   prob = float(input(f"Enter the probability {newsday}: "))
   probability.append(prob)


   temp += prob
   cmltv_prob.append(temp)




demand = []
good_prob = []
fair_prob = []
poor_prob = []
good_cmltv_prob = []
fair_cmltv_prob = []
poor_cmltv_prob = []
gpp=0
fpp=0
ppp=0
d=int(input("Number of demand : "))
for i in range((d)):
   temp = input(f"Enter the demand {i+ 1}: ")
   demand.append(temp)

   gp = float(input(f"Enter the probability for good {i+ 1}: "))
   good_prob.append(gp)

   fp = float(input(f"Enter the probability for fair {i+ 1}: "))
   fair_prob.append(fp)

   pp = float(input(f"Enter the probability for poor {i+ 1}: "))
   poor_prob.append(pp)
   gpp += gp
   good_cmltv_prob.append(gpp)
   fpp+= fp
   fair_cmltv_prob.append(fpp)

   ppp+= pp
   poor_cmltv_prob.append(ppp)


Newsday_rand=[]
demand_rand=[]
temp = 0
Xn = []
a=3
seed=5
c=3
m=7
i=0
day1=int(input("How many day : "))
for i in range((day1)):
    #temp = input(f"Enter the random digit for types of newsday {i+ 1}: ")
    #print(f"Enter the random digit for types of newsday {i+ 1}: ")
    seed = (a * seed + c) % m
    Xn.append(seed)


    r= seed/m

    Newsday_rand.append(int(r*100))

Xd = []
a=13
seed=3
c=0
m=64
for i in range((day1)):
    #temp = input(f"Enter the random digit for demand {i+ 1}: ")
    #print(f"Enter the random digit for demand {i+ 1}: ")
    seed = (a * seed + c) % m
    Xd.append(seed)


    r= seed/m
    demand_rand.append(int(r*100))


def types_newsday(value):
    value = int(value)
    if value >= 0 and value <= 35:
        return "Good"
    elif value >= 36 and value <= 80:
        return "Fair"
    elif value >= 81 and value <= 100:
        return "Poor"

def good(value):
    value = int(value)
    if value >= 0 and value <= 3:
        return 40
    elif value >= 4 and value <= 8:
        return 50
    elif value >= 9 and value <= 23:
        return 60
    elif value >= 24 and value <= 43:
        return 70
    elif value >= 44 and value <= 78:
        return 80
    elif value >= 79 and value <= 93:
        return 90
    elif value >= 94 and value <= 100:
        return 100

def fair(value):
    value = int(value)
    if value >= 0 and value <= 10:
        return 40
    elif value >= 11 and value <= 28:
        return 50
    elif value >= 29 and value <= 68:
        return 60
    elif value >= 69 and value <= 88:
        return 70
    elif value >= 89 and value <= 96:
        return 80
    elif value >= 97 and value <= 100:
        return 90


def poor(value):
    value = int(value)
    if value >= 0 and value <= 44:
        return 40
    elif value >= 45 and value <= 66:
        return 50
    elif value >= 67 and value <= 82:
        return 60
    elif value >= 83 and value <= 94:
        return 70
    elif value >= 95 and value <= 100:
        return 80



day = []
types_of_newsday = []
demand1 = []
revenue_from_sales=[]
lost_profit=[]
salvage=[]
daily_profit=[]
temp = 0
buy=int(input("Enter purchased : "))
stock=int(input("Enter stock : "))
selling=int(input("Selling Price : "))
de=int(input("Enter demand value : "))
scrap=int(input("Enter scrap value : "))
cost_of_newsday=(stock*buy)/100
i=0
total_salvage=0
total_daily_profit=0
total_revenue=0
total_lost_profit=0
for i in range((day1)):
   day.append(i+1)
   types_of_newsday.append(types_newsday(Newsday_rand[i]))
   if( types_of_newsday[i]=="Good"):
          demand1.append(good(demand_rand[i]))
   elif( types_of_newsday[i]=="Fair"):
          demand1.append(fair(demand_rand[i]))
   elif( types_of_newsday[i]=="Poor"):
          demand1.append(poor(demand_rand[i]))
   if(demand1[i]<stock):
          revenue = demand1[i]*(selling/100)
   elif(demand1[i]>=stock):
          revenue = stock*(selling/100)
   total_revenue+=revenue
   revenue_from_sales.append("${:.2f}".format(revenue))
   if(demand1[i]>stock):
          lost=(demand1[i]-stock)*de/100
          lost_profit.append("${:.2f}".format(lost))
          salvage.append("-")
          daily=revenue - cost_of_newsday - lost
          daily_profit.append("${:.2f}".format(daily))
          total_lost_profit+=lost

   elif (demand1[i]<stock):
          sal=(stock-demand1[i])*scrap/100
          lost_profit.append("-")
          salvage.append("${:.2f}".format(sal))
          daily =revenue - cost_of_newsday + sal
          daily_profit.append("${:.2f}".format(daily))
          total_salvage+=sal

   elif(demand1[i]==stock) :
          lost_profit.append("-")
          salvage.append("-")
          daily=revenue - cost_of_newsday
          daily_profit.append("${:.2f}".format(daily))


   total_daily_profit+= daily

print("\nDistribution Table for Type of Newsday : ")
data = list(zip(Newsdays, probability, cmltv_prob))
headers = ["Newsdays", "Probability"," Cumulative Probability"]
table = tabulate(data, headers, tablefmt="fancy_grid")
print(table)

print("\nDistribution Table for Newspapers Demanded : ")
data = list(zip(demand, good_prob, good_cmltv_prob,fair_prob,fair_cmltv_prob,poor_prob, poor_cmltv_prob))
headers = ["Demand", "good_prob","good_cmltv_prob","fair_prob","fair_cmltv_prob","poor_prob", "poor_cmltv_prob"]
table = tabulate(data, headers, tablefmt="fancy_grid")
print(table)

print("\nSequence : ")
data = list(zip(Xn,Xd))
headers = ["Type of Newsday-Sequence (Xn)","Newspapers Demanded-Sequence (Xd)"]
table = tabulate(data, headers, tablefmt="fancy_grid")
print(table)


print("\nSimulation Table :")


data = list(zip(day, Newsday_rand, types_of_newsday, demand_rand, demand1, revenue_from_sales, lost_profit, salvage, daily_profit))
headers = ["Day", "RD newsday","Types of newsday", "RD demand", "Demand", "Revenue from sales", "lost profit", "Salvage from sales", "Daily profit"]
total_row = ["Total", "-", "-", "-", "-", "${:,.2f}".format(total_revenue), "${:,.2f}".format(total_lost_profit),
             "${:,.2f}".format(total_salvage), "${:,.2f}".format(total_daily_profit)]
data.append(total_row)
table = tabulate(data, headers, tablefmt="fancy_grid")
print(table)
```
